```python
import glob
from tqdm.notebook import tqdm_notebook
networks = []

for instance_path in tqdm_notebook(glob.glob('cell-instance-segmentation/sample_input/*.data'), 
                                   desc = instance_path[40:]):
    instances = {}
    with open(instance_path, 'rb') as filehandle:
        try:
            instances = pickle.load(filehandle)
            networks.append(network_object(instances = instances, adjacency_list = None, 
                                   tag = instance_path[45:47]))
        except RuntimeError:
            #instances = torch.load(filehandle, pickle_module=pickle, map_location=torch.device('cpu'))
            print("load failed for "+instance_path)
```

```python
rand_network = [network_object(instances = None, adjacency_list = None, tag = None, mode='rand') for i in range(5)]
bary_network = [network_object(instances = None, adjacency_list = None, tag = None, mode='bary') for i in range(5)]
```

```python
e14 = [net for net in networks if net.tag == '14']
e15 = [net for net in networks if net.tag == '15']
e16 = [net for net in networks if net.tag == '16']
e17 = [net for net in networks if net.tag == '17']
e18 = [net for net in networks if net.tag == '18']
ezz = [rand_network,e14,e15,e16,e17,e18,bary_network]
```

```python
avg = []
var = []
ent = []
ass = []

for e in ezz:
    temp_avg = []
    temp_var = []
    temp_ent = []
    temp_ass = []
    for net in e:
        temp_avg.append(net.average_degree())
        temp_var.append(net.degree_variance())
        temp_ent.append(net.degree_entropy())
        temp_ass.append(net.assortativity())
    avg.append(temp_avg)
    var.append(temp_var)
    ent.append(temp_ent)
    ass.append(temp_ass)
```

```python
def custom_boxplot(thing, num_sets):
    plt.plot(range(1,num_sets+1),[sum(thing[0])/len(thing[0]) for i in range(num_sets)])
    plt.plot(range(1,num_sets+1),[sum(thing[-1])/len(thing[-1]) for i in range(num_sets)])
    plt.boxplot(thing, notch=True)
    # x = [item+(random.random()-.5)*.2 for sublist in range(len(thing)) for item in sublist]
    
    x = [[i for j in range(len(thing[i]))] for i in range(len(thing))]
    x = [1+item+(random.random()-.5)*.2 for sublist in x for item in sublist]
    
    y = [item for sublist in thing for item in sublist]
    plt.plot(x, y, 'r.', alpha=0.5)
```

```python
custom_boxplot(avg, 7)
plt.title('average neighbors')
plt.legend(['Random', 'HCP'])
plt.show()
custom_boxplot(var, 7)
plt.title('variance in neighbors')
plt.legend(['Random', 'HCP'])
plt.show()
custom_boxplot(ent, 7)
plt.title('entropy in neighbors')
plt.legend(['Random', 'HCP'])
plt.show()
custom_boxplot(ass, 7)
plt.title('assortativity')
plt.legend(['Random', 'HCP'])
plt.show()
```
