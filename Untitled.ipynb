```python
from scipy.spatial import Voronoi, voronoi_plot_2d, ConvexHull
import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.patches as patches
import networkx as nx
from scipy.stats import norm
import seaborn as sns

from network_object import network_object
```

```python
globalmin = -1
globalmax = 1

points = np.array([[random.random(),random.random()] for i in range(500)])
points = points*(globalmax-globalmin)+globalmin
rand_v = Voronoi(points)
# cart_v = Voronoi(cart_coords)
fig = voronoi_plot_2d(rand_v)
plt.show()
# fig = voronoi_plot_2d(cart_v)
# plt.show()
```

```python
rand_v.regions = [r for r in rand_v.regions if len(r)!=0]
polygons = [([x[0] for x in polygon],[y[1] for y in polygon]) 
            for polygon in [[rand_v.vertices[point] for point in region] for region in rand_v.regions]]
polygons = [[rand_v.vertices[point] for point in region] for region in rand_v.regions]

buffer = (globalmax-globalmin)*0.05
polygons = [polygon for polygon in polygons if np.amax(polygon)<(globalmax-buffer) 
                                            and np.amin(polygon)>(globalmin+buffer)]

```

```python
xs = [[rand_v.vertices[i[j]][0] for j in range(len(i))] for i in rand_v.regions]
ys = [[rand_v.vertices[i[j]][1] for j in range(len(i))] for i in rand_v.regions]

buffer = (globalmax-globalmin)*0.05
rand_v.regions = [rand_v.regions[i] for i in range(len(rand_v.regions)) if np.amax(xs[i])<(globalmax-buffer)
                                            and np.amin(xs[i])>(globalmin+buffer)
                                            and np.amax(ys[i])<(globalmax-buffer)
                                            and np.amin(ys[i])>(globalmin+buffer)]
```

```python
shapes = []
maxlen = max([len(poly) for poly in polygons])+1

for poly in polygons:
    color = ((len(poly)-1)/maxlen, (len(poly)-1)/maxlen, (len(poly)-1)/maxlen)
    
    shapes.append(patches.Polygon(poly, facecolor=color, edgecolor=(0,0,0)))
```

```python
fig, ax = plt.subplots()
ax.set(xlim=(globalmin, globalmax), ylim=(globalmin, globalmax))

for sh in shapes: 
    ax.add_patch(sh)
```

```python
hulls = [ConvexHull(poly) for poly in polygons]
num_neighbors = [len(hull.neighbors) for hull in hulls]
areas = [hull.area for hull in hulls]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,3))

mu, std = norm.fit(areas)
x = np.linspace(0, 1, 100)
p = norm.pdf(x, mu, std)
ax1.plot(x, p, 'k', linewidth=2)
ax1.hist(areas, bins = 100)
ax1.set_title('mu = {}, std = {}'.format(mu,std))

mu, std = norm.fit(num_neighbors)
x = np.linspace(2, 11, 100)
p = norm.pdf(x, mu, std)
ax2.plot(x, p*300, 'k', linewidth=2)
ax2.hist(num_neighbors, bins = np.unique(num_neighbors))
ax2.set_title('mu = {}, std = {}'.format(mu,std))
plt.show()
```
